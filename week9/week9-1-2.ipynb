# 순환 신경망으로 IMDB 리뷰 분류하기

## IMDB 리뷰 데이터셋

```python
 from tensorflow.keras.datasets import imdb
```

```python
(train_input, train_target), (test_input, test_target) = imdb.load_data(num_words=500)
```

```python
print(train_input.shape, train_target.shape)
print(test_input.shape, test_target.shape)
```

```python
print(len(train_input[0]))
```

```python
print(len(train_input[1]))
```

```python
# 리뷰가 긍정인지 부정인지 판단
# 0은 부정, 1은 긍정
print(train_target[:20])
```

```python
from sklearn.model_selection import train_test_split

train_input, val_input, train_target, val_target = train_test_split(
    train_input, train_target, test_size=0.2, random_state=42
)
```

```python
print(train_input.shape, train_target.shape)
print(val_input.shape, val_target.shape)
```

```python
# 각 리뷰의 길이를 계산해서 넘파이 배열에 추가
import numpy as np

lengths = np.array([len(x) for x in train_input])
```

```python
# 리뷰 길이의 평균과 중간값 출력
print(np.mean(lengths), np.median(lengths))
```

```python
# 히스토그램으로 표현
import matplotlib.pyplot as plt

plt.hist(lengths)
plt.xlabel("lengths")
plt.ylabel("frequency")
plt.show()
```

```python
# 케라스는 시퀀스 데이터의 길이를 맞추는 pad_sequences() 함수를 제공
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
train_seq = pad_sequences(train_input, maxlen=100)
val_seq = pad_sequences(val_input, maxlen=100)

print(train_seq.shape) # 파이썬 리스트의 배열이 2차원 배열이 됨
print(val_seq.shape)
```

```python
# 패딩값인 0이 없으므로 원본 데이터는 100보다 길었을 것
print(train_seq[0])
```

```python
# 원본 데이터의 끝에서 10개 토큰 확인
# 일반적으로 시퀀스의 뒷부분의 정보가 더 유용하리라 기대하기 때문에 maxlen보다 킨 시퀀스의 앞부분을 잘라낸다.
print(train_input[0][-10:]) 
```

```python
# 6번째 샘플 출력
# 패딩 토큰은 샘플의 앞부분에 추가된다.
print(train_seq[5])
```

## 순환 신경망 만들기

```python
from tensorflow import keras
```

```python
model = keras.Sequential()
model.add(keras.layers.SimpleRNN(8, input_shape=(100, 500)))
model.add(keras.layers.Dense(1, activation="sigmoid"))
```

```python
# 정수 토큰값을 길이 500인 원-핫 인코딩 배열로 변환
train_oh = keras.utils.to_categorical(train_seq)
val_oh = keras.utils.to_categorical(val_seq)

print(train_oh.shape)
print(val_oh.shape)
```

```python
# 첫 번째 샘플의 첫 번째 토큰(10) 출력
print(train_oh[0][0])
print(np.sum(train_oh[0][0]))
```

```python
# 모델의 구조 출력
model.summary()
```

## 순환 신경망 훈련하기

```python
rmsprop = keras.optimizers.RMSprop(learning_rate=1e-4)
model.compile(optimizer=rmsprop, loss="binary_crossentropy", metrics=["accuracy"])

checkpoint_cb = keras.callbacks.ModelCheckpoint("best-simplernn-model.h5", save_best_only=True)
early_stopping_cb = keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True)

history = model.fit(
    train_oh, train_target, 
    epochs=100, batch_size=64, 
    validation_data=(val_oh, val_target), 
    callbacks=[checkpoint_cb, early_stopping_cb]
)
```

```python
# 훈련 손실과 검증 손실을 그래프로 그리기
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend(["train", "val"])
plt.show()
```

```python
# train_seq 배열과 train_oh 배열의 크기를 비교
print(train_seq.nbytes, train_oh.nbytes)
```

## 단어 임베딩을 사용하기

```python
model2 = keras.Sequential()
model2.add(keras.layers.Embedding(500, 16, input_shape=(100,)))
model2.add(keras.layers.SimpleRNN(8))
model2.add(keras.layers.Dense(1, activation="sigmoid"))
```

```python
model2.summary()
```

```python
rmsprop = keras.optimizers.RMSprop(learning_rate=1e-4)
model2.compile(optimizer=rmsprop, loss="binary_crossentropy", metrics=["accuracy"])

checkpoint_cb = keras.callbacks.ModelCheckpoint("best-simplernn-model.h5", save_best_only=True)
early_stopping_cb = keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True)

history = model2.fit(
    train_seq, train_target, 
    epochs=100, 
    batch_size=64, 
    validation_data=[val_seq, val_target], 
    callbacks=[checkpoint_cb, early_stopping_cb]
)
```

```python
# 훈련 손실과 검증 손실을 그래프로 그리기
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend(["train", "val"])
plt.show()
```
